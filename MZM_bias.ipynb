```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 21 13:19:13 2021

Modulator position DC

@author: dmoreno
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from __future__ import print_function
from ipywidgets import interact, interactive, fixed

plt.rcParams['figure.figsize'] = [26, 13]
import plotly.graph_objects as go
import plotly.express as px

@interact(Offset=(0,12.2,0.1),Vpi=(0,7,0.1),ExtraIL=(-10,10,1),lim_superior=3.73, lim_inferior=-2.6)
def bias(Offset,Vpi, ExtraIL, lim_superior, lim_inferior):
  Nae = 8                            # Number of antenna elements

  frf1 = 29.0*1e9 
  frf2  = 9.8*1e9                 # Frequency 2               [GHz]
  fbase = 0.1e9                      # ¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿
  Nfft  = 2**12                      # Number of symbols
  fs    = Nfft*fbase                 # Define sampling frequency [Hz]
  Nfft  = Nfft*(2**7)                 # Number of samples
  Ts    = 1/fs                       # Sampling period           [s]
  tvec  = np.arange(0,(Nfft-1)*Ts+Ts,Ts)           # Time vector    
  fd    = np.linspace(-0.5, 0.5, Nfft)  # Discrete Frequency        [Hz]
  f     = 2*fbase
  fvec  = fd*fs                      # Frequency vector          [Hz]
  freq  = fvec*1e-9
  plots_on = 1


  Pin_dB=14
  Pin=10**((Pin_dB-30)/10)
  Impedance=50
  Vin=np.sqrt(2* Impedance * Pin)

  frf=1

  #Vpi=3.5
  IL=10**((8.6+20+3+ExtraIL)/10)
  # IL=IL+1

  ##--------Responsividad------------------------------
  Resp=0.9

  ''''Rapida representacion'''
  VV=np.linspace(-20,30,50000)
  TV=0.5*(1+np.cos(np.pi*(VV)/Vpi))/IL
  TV_dB=10*np.log10(0.5*(1+np.cos(np.pi*VV/Vpi))/IL)
  #Crossing points
  TV_373=10*np.log10(0.5*(1+np.cos(np.pi*(3.73-Offset)/Vpi))/IL)
  TV_26=10*np.log10(0.5*(1+np.cos(np.pi*(-2.6-Offset)/Vpi))/IL)


  #---------Función de transferencia------------------
  # plt.figure()
  # plt.plot(VV,TV_dB)
  # plt.grid(True)
  # plt.title("T(V)")

  #---------Potencia de salida------------------------

  Pout_dB=TV_dB+Pin_dB
  Pout_373=TV_373+Pin_dB
  Pout_26=TV_26+Pin_dB
  fig = go.Figure(data=go.Scatter(x=VV+Offset, y=Pout_dB, name="Pout"))
  fig.add_trace(go.Scatter(x=[lim_inferior, lim_inferior], y=[-300,300], mode="lines", name="Limite inferior"))
  fig.add_trace(go.Scatter(x=[lim_superior, lim_superior], y=[-300,300], mode="lines", name="Limite superior"))
  
  fig.update_layout(xaxis=dict(title='Vbias (V)'), xaxis_range=[-5,5], yaxis=dict(title='Pout (dBm)'),yaxis_range=[-40,-15], template="plotly_white", title="MZM", autosize=False, width=920, height=500)
  fig.show()
```
